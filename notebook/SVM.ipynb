```python
import cv2 as cv
import numpy as np
import pandas as pd
import os 
```

```python
keypoints = []
descriptors = []
folders = os.listdir("../Train/")
for folder in folders:
    files = os.listdir("../Train/" + folder)
    for file in files:
        path = '../Train/'+ folder + '/' + file
        img = cv.imread(path)
        img = cv.resize(img, (224, 224))
        sift = cv.xfeatures2d.SIFT_create()
        kp = sift.detect(img,None)

        img=cv.drawKeypoints(img,kp, outImage = None)

        kp1, des1 = sift.detectAndCompute(img,None)
        print(des1.shape)
        keypoints.append(kp1)
        descriptors.append(des1)
        
    
data = {'keypoints': keypoints,
        'descriptors': descriptors}
df = pd.DataFrame(data, columns = ['keypoints','descriptors'])
df.to_excel('../descriptors.xlsx', index = False)
```
