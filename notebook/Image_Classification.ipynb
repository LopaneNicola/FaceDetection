```python
import numpy as np
import cv2
import os
import pandas as pd
```

```python
data = pd.read_csv('../data/fer2013.csv')
train_set = data[(data.Usage == 'Training')]
validation_set = data[(data.Usage == 'PublicTest')]
test_set = data[(data.Usage == 'PrivateTest')]
```

```python
print(len(train_set))
print(len(validation_set))
print(len(test_set))
```
